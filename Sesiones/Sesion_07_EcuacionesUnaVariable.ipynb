```python
import numpy as np
import matplotlib.pylab as plt
```

```python
f = lambda theta: theta - np.sin(theta) - 0.5 * np.pi
```

```python
theta = np.linspace(0, 2*np.pi, 100)

plt.figure(figsize=(6,4))
plt.plot(theta, f(theta), label="$f(\\theta)$")
plt.xlabel("$\\theta$")
plt.ylabel("$f(\\theta)$")
plt.grid()
plt.legend()
plt.show()
```

```python
from scipy import optimize
```

```python
theta = optimize.bisect(f, a=1, b=6, xtol=1E-12)
print(f"la raiz de f(theta) es {theta:.7}")
```

```python
theta = optimize.bisect(f, a=1, b=6, xtol=1E-12)
print(f"la raiz de f(theta) es {theta:.8}")
```
